```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('KNN_Project_Data')
```

```python
df.head()
```

```python
sns.pairplot(df, hue='TARGET CLASS', palette='husl')
```

# Standardizing Variables

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df.drop('TARGET CLASS', axis=1))
```

```python
scaler_features = scaler.transform(df.drop('TARGET CLASS', axis=1))
```

```python
df_feature = pd.DataFrame(scaler_features, columns=df.columns[:-1])
```

```python
df_feature.head()
```

# Tain Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_feature
y = df['TARGET CLASS']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
pred = knn.predict(X_test)
```

```python
print(classification_report(y_test, pred))
```

```python
print(confusion_matrix(y_test, pred))
```

# Optimal K Value

```python
error = []

for i in range(1,60):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    error.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(12,5))

plt.plot(range(1,60), error, color='red', ls='--', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Error rate vs K Value')
plt.xlabel('K Values')
plt.ylabel('Error Rate')
```

# Retraining with New K Value

```python
knn = KNeighborsClassifier(n_neighbors=31)

knn.fit(X_train, y_train)

pred = knn.predict(X_test)
print(classification_report(y_test, pred))
print(confusion_matrix(y_test, pred))
```
